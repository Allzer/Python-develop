Делаем линейную регрессию

```python
k = 2.8
b = -3
x = np.linspace(1,100, 100).reshape(-1,1).astype('int32')
y = [i*k for i in range(100)]
y = np.array(y).reshape(-1,1).astype('int32')
```

```python
model = LinearRegression().fit(x, y)
y_p = model.predict(x)
```

```python
plt.scatter(x,y, s = 10)
plt.plot(x,y_p, c ='r')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
